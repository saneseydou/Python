```python
import numpy as np

grades=np.array([85, 90, 88, 92, 95, 80, 75, 98, 89, 83]) #soit le tablau à une seule ligne
print(grades)

print(f"la moyenne de grades est : {np.mean(grades)}") #le calcul de la moyenne de toute le tablau

print(f"la médiane de grades est : {np.median(grades)}") #calcul de la médiane

print(f"la médiane de grades est : {np.std(grades)}") #calcul de l'écart type (standard deviation)

print(f"le maximum de grades est : {np.max(grades)}") #trier le maximum

print(f"le minimum de grades est : {np.min(grades)}") #trier le minimum


y=np.sort(grades)                                 #classement dans l'ordre croissant
print(f"arragement dans l'ordre croisssant : {y}")


print(f" les individus ayant des notes supérieur à 90 sont : {np.count_nonzero(grades > 90)}") #faire une décompte


print(f" les individus ayant des notes supérieur à 90 sont : {np.count_nonzero(grades > 90)}") #calcul du pour centage de notes superieur à 90

meilleur= np.argmax([grades])
print(meilleur)
```
