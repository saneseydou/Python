```python
import numpy as np

grades=np.array([85, 90, 88, 92, 95, 80, 75, 98, 89, 83]) #soit le tablau à une seule ligne
print(grades)

print(f"la moyenne de grades est : {np.mean(grades)}") #le calcul de la moyenne de toute le tablau

print(f"la médiane de grades est : {np.median(grades)}") #calcul de la médiane

print(f"la médiane de grades est : {np.std(grades)}") #calcul de l'écart type (standard deviation)

print(f"le maximum de grades est : {np.max(grades)}") #trier le maximum

print(f"le minimum de grades est : {np.min(grades)}") #trier le minimum


y=np.sort(grades)                                 #classement dans l'ordre croissant
print(f"arragement dans l'ordre croisssant : {y}")


print(f" les individus ayant des notes supérieur à 90 sont : {np.count_nonzero(grades > 90)}") #faire une décompte

#calcul du pourcentage de notes superieur à 90
x=(np.count_nonzero(grades > 90))*100      #le nbre de notes sup à 90 *100
y=np.count_nonzero(grades)                  #le nbre total de notes existant
pourcentage_1= x/y
print(f"le pourcentage des notes supérieur à 90 est : {pourcentage_1} %")

#calcul du pourcentage de notes inférieur à 75

z=(np.count_nonzero(grades< 75))*100   #le nbre de notes sup à 75 *100
pourcentage_2= z/y
print(f"le pourcentage des notes inférieur à 75 est : {pourcentage_2} %")


A= np.argmax(grades, axis=0)  #argmax nous permet d'indexer la meilleure note avec axis=0 qui indique les lignes
print(A)

high_performers=np.array(grades[grades>90])     #le [grades>90] nous sort des booleen false and true et le premier grades nous extrait pour sortir des valeurs

print(f" votre tableau des valeurs superieur à 90 est :\n {high_performers}")


passing_grades=np.array(grades[grades< 75])                                     #le [grades>75] nous sort des booleen false and true et le premier grades nous extrait pour sortir des valeurs
print(f" votre tableau des valeurs inférieur à 75 est :\n {passing_grades}")

#------------------------------------------------------------------------------------------------------

# Votre Objectif

# Créez un tableau numpy appelé « notes » qui contient les notes suivantes : [85, 90, 88, 92, 95, 80, 75, 98, 89, 83]


# Instructions

# Créez un nouveau fichier appelé « grades_analysis.py »
# Importez la bibliothèque numpy et créez le tableau « grades » comme spécifié ci-dessus.
# Utilisez les fonctions numpy pour calculer la moyenne, la médiane et l’écart type des notes.
# Utilisez la fonction numpy pour trouver le maximum et le minimum des notes.
# Utilisez la fonction numpy pour trier les notes par ordre croissant.
# Utilisez la fonction numpy pour trouver l'index de la note la plus élevée dans le tableau.
# Utilisez la fonction numpy pour compter le nombre d'étudiants ayant obtenu un score supérieur à 90.
# Utilisez la fonction numpy pour calculer le pourcentage d'étudiants ayant obtenu un score supérieur à 90.
# Utilisez la fonction numpy pour calculer le pourcentage d'étudiants ayant obtenu un score inférieur à 75.
# Utilisez la fonction numpy pour extraire toutes les notes supérieures à 90 et les placer dans un nouveau tableau appelé « high_performers ».
# Créez un nouveau tableau appelé « passing_grades » qui contient toutes les notes supérieures à 75.
# Imprimez le résultat de toutes les étapes ci-dessus.
# Note:

# pour calculer le pourcentage, utilisez numpy.mean(grades > 90) * 100
# pour extraire les notes supérieures à 90, utilisez notes[notes > 90]
# pour extraire les notes supérieures à 75, utilisez notes[notes > 75]
# Vous pouvez également utiliser d'autres fonctions numpy pour analyser les données comme vous le souhaitez. Les étapes ci-dessus ne sont que des exemples de ce qui peut être fait.
```
